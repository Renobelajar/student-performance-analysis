1. Menampilkan Dataset

```python
import pandas as pd

# Membaca data
df = pd.read_csv('data_pelatihan.csv')

# Menampilkan data
df.head(15) 
```

2. Nilai Uts Tertinggi dan Terendah


```python
tertinggi_uts = df['UTS'].max()
terendah_uts = df['UTS'].min()

print(f"Nilai Tertinggi: {tertinggi_uts}")
print(f"Nilai Terendah: {terendah_uts}")
```

3. Nilai Uas Tertinggi dan Terendah

```python
tertinggi_uas = df['UAS'].max()
terendah_uas = df['UAS'].min()

print(f"Nilai Tertinggi: {tertinggi_uas}")
print(f"Nilai Terendah: {terendah_uas}")
```

4. Total Nilai Tertinggi dan Terendah

```python
tertinggi = df['Nilai'].max()
terendah = df['Nilai'].min()

print(f"Nilai Tertinggi: {tertinggi}")
print(f"Nilai Terendah: {terendah}")
```

5. Nilai Laki - laki dan Perempuan tertinggi

```python
# 1. Cari indeks baris yang nilainya paling tinggi
indeks_max = df['Nilai'].idxmax()
indeks_min = df['Nilai'].idxmin()

# 2. Ambil nama berdasarkan indeks tersebut
nama_tertinggi = df.loc[indeks_max, 'Nama']
nama_terendah = df.loc[indeks_min, 'Nama']

print(f"Nilai tertinggi diraih oleh: {nama_tertinggi}")
print(f"Nilai terendah diraih oleh: {nama_terendah}")
```

6. Rata - rata nilai keseluruhan siswa

```python
ratarata = df['Nilai'].mean()

print(f"Nilai Rata rata: {ratarata}")
```

7. Rata - rata Nilai laki - laki dan perempuan

```python
# Rata-rata Laki-laki
rata_laki = df[df['Gender'] == 'Laki-laki']['Nilai'].mean()

# Rata-rata Perempuan
rata_perempuan = df[df['Gender'] == 'Perempuan']['Nilai'].mean()

print(f"Nilai Rata-rata laki-laki: {rata_laki}")
print(f"Nilai Rata-rata Perempuan: {rata_perempuan}")
```

8. Matkul dengan nilai tertinggi 

```python
nilai_tertinggi = df.loc[df['Nilai'].idxmax()]

# Menampilkan hasil
print("Mata kuliah dengan nilai tertinggi:")
print(f"Mata Kuliah: {nilai_tertinggi['Matkul']}")
print(f"Nilai: {nilai_tertinggi['Nilai']}")
```

9. Nilai Uts Tertinggi dan Terendah dengan grafik batang


```python
import matplotlib.pyplot as plt
tertinggi_uts = 95  # Nilai UTS tertinggi
terendah_uts = 45   # Nilai UTS terendah

labels = ['Tertinggi', 'Terendah']
values = [tertinggi_uts, terendah_uts]

# Membuat diagram batang
plt.bar(labels, values, color=['green', 'blue'])

# Menambah judul dan label
plt.title('Perbandingan Nilai UTS Tertinggi vs Terendah')
plt.ylabel('Nilai')
plt.show()
```

10. Nilai Uas Tertinggi dan Terendah dengan grafik batang



```python
tertinggi_uas = 95  # Nilai UAS tertinggi
terendah_uas = 65   # Nilai UAS terendah

labels = ['Tertinggi', 'Terendah']
values = [tertinggi_uas, terendah_uas]

# Membuat diagram batang
plt.bar(labels, values, color=['green', 'blue'])

# Menambah judul dan label
plt.title('Perbandingan Nilai UTS Tertinggi vs Terendah')
plt.ylabel('Nilai')
plt.show()
```

11. Nilai Tertinggi per matkul dengan grafik batang

```python
# Mencari nilai tertinggi per mata kuliah
nilai_tertinggi_per_matkul = df.groupby('Matkul')['Nilai'].max()

# Menampilkan hasil dengan format yang lebih rapi
print("Nilai tertinggi per mata kuliah:")
for matkul, nilai in nilai_tertinggi_per_matkul.items():
    print(f"{matkul}: {nilai}")

# Membuat grafik batang
nilai_tertinggi_per_matkul.plot(kind='bar', color='skyblue', figsize=(8, 5))

# Menambahkan judul dan label
plt.title('Nilai Tertinggi per Mata Kuliah')
plt.xlabel('Mata Kuliah')
plt.ylabel('Nilai Tertinggi')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

Dengan diagram ling

```python
import matplotlib.pyplot as plt

# Mencari nilai tertinggi per mata kuliah
nilai_tertinggi_per_matkul = df.groupby('Matkul')['Nilai'].max()

print("Nilai tertinggi per mata kuliah:")
for matkul, nilai in nilai_tertinggi_per_matkul.items():
    print(f"{matkul}: {nilai}")

# Membuat diagram lingkaran
fig, ax = plt.subplots(figsize=(8, 8))
nilai_tertinggi_per_matkul.plot(
    kind='pie', 
    autopct='%1.1f%%',  
    startangle=90,      
    colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'],  
    legend=True,
    ax=ax  # Menggunakan objek ax
)

# Menambahkan judul
ax.set_title('Persentase Nilai Tertinggi per Mata Kuliah')

# Menghapus label sumbu y
ax.set_ylabel('')

# Menampilkan diagram lingkaran
plt.show()
```

Versi donat

```python


# Mencari nilai tertinggi per mata kuliah
nilai_tertinggi_per_matkul = df.groupby('Matkul')['Nilai'].max()

print("Nilai tertinggi per mata kuliah:")
for matkul, nilai in nilai_tertinggi_per_matkul.items():
    print(f"{matkul}: {nilai}")

# Membuat diagram donat
fig, ax = plt.subplots(figsize=(8, 8))
nilai_tertinggi_per_matkul.plot(
    kind='pie', 
    autopct='%1.1f%%',  
    startangle=90,      
    colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'],  
    legend=True,
    ax=ax,  # Menggunakan objek ax
    wedgeprops={'width': 0.3}  # Membuat lubang di tengah
)

# Menambahkan judul
ax.set_title('Persentase Nilai Tertinggi per Mata Kuliah')

# Menghapus label sumbu y
ax.set_ylabel('')

# Menampilkan diagram donat
plt.show()
```
