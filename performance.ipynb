2. Uts Tertinggi dan Terendah


```python
import matplotlib.pyplot as plt

# Asumsi data Anda:
kategori = ['A', 'B', 'C', 'D', 'E']
nilai = [45, 82, 23, 95, 60] # Ganti dengan data Anda

# 1. Cari nilai ekstrem
maks = max(nilai)
mins = min(nilai)

# 2. Buat daftar warna berdasarkan kondisi
warna = []
for x in nilai:
    if x == maks:
        warna.append('forestgreen') # Hijau untuk tertinggi
    elif x == mins:
        warna.append('crimson')     # Merah untuk terendah
    else:
        warna.append('lightgrey')   # Abu-abu untuk sisanya

# 3. Plot
plt.bar(kategori, nilai, color=warna)

# Opsional: Tambah teks agar angkanya langsung terlihat
plt.text(nilai.index(maks), maks + 1, f'Max: {maks}', ha='center', color='green', fontweight='bold')
plt.text(nilai.index(mins), mins + 1, f'Min: {mins}', ha='center', color='red', fontweight='bold')

plt.title('Perbandingan Tertinggi vs Terendah')
plt.show()
```
